<b>Descripción:</b>
Los sistemas de recomendación son soluciones informáticas capaces de brindar al usuario información de su preferencia o de su interés. Por ejemplo, la recomendación de películas o canciones que nos realizan las plataformas de streaming, se basan en esta tecnología.
Usted esta desarrollando una aplicación para gestionar y recomendar compras par el hogar. Respecto a la recomendación de compras la plataforma debería sugerir elementos a los usuarios de acuerdo a sus características y la eleccion que han hecho históricamente otros usuarios. Por ejemplo, si un usuario ha comprado vegetales y es vegetariano, el sistema debería recomendarle que comprara Tofu, debido a que este último ítem aparece frecuentemente en usuarios con características y preferencias similares.
Para este trabajo nos enfocaremos en una primera etapa que consiste en definir una distancia entre dos usuarios (U1, U2). Esta distancia debería ser alta si los usuarios tienen características y preferencias similares, y baja en el caso contrario.
Para ello usted diseñará y realizará una prueba de concepto utilizando lógica borrosa. A continuación unas preguntas y actividades guía para realizar este prototipo:
</p>

<b>
3. ¿Cómo representará los elementos de preferencia de cada usuario? Un vector de elementos? Un vector de categorías de elementos? Que conjuntos difusos pueden ser establecidos para esta entrada?
</b>

<ul>
    <li>
 Se representará como un vector de categorías de elementos porque de esta manera se podrán realizar la selección de categoria(s)  a recomendar a dicho usuario segun sus caracteriticas(o el complimiento de las reglas) y la precisión en los productos a recomendar dependerá de que tan grueso o fino sea la categorización de los productos de Hogar. Por ejemplo en la tabla anterior se listaron 10 categorias pero si queremos que la categorización sea mas fina como el ejemplo de que si es vegetariano entonces recomiende tofu entonces podriamos incrementar el número de categorias hasta el triple, cuádruple o más. Ya que por ejemplo la categoria carnes tiene carnes frias, carnes rojas, carnes blancas entre otras.
        
    </li>
    <br>
    <li>
    Para esta entrada los conjuntos difusos como anteriormente se mencionó no son mas que las categorias de productos que vayamos a definir en este caso las 10 categorías listadas en la tabla anterior.
    </li>
</ul>
<br>
<b>
4. ¿Cómo representará la distancia entre los dos usuarios U1 y U2? Que conjuntos difusos pueden ser establecidos para esta salida?
</b>
<br>
Para representar en terminos de distancia la diferencia entre un usuario y otro, se va a determinar que cada usuario es un vector de caracteristicas y que el vector resultante de la resta de ambos es el vector que contiene la diferencia (distancia) los dos.
